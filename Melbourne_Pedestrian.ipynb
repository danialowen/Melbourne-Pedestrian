## Chaging the Date_Time


```python
df.DateTime = df.loc[df['Sensor_ID'] == 1]
```

```python
df.DateTime.head()
```

```python
df.DateTime1 = df.DateTime[['Date_Time']]
```

```python
df.DateTime1.head()
```

```python
df.DateTime1['Date_Time'] = pd.to_datetime(df.DateTime1['Date_Time'])
```

```python
df.DateTime1.head(14)
```

```python
#Create new variables of date time and hour
df.DateTime1['Year'] = df.DateTime1['Date_Time'].dt.year
df.DateTime1['Month'] = df.DateTime1['Date_Time'].dt.month
df.DateTime1['Day'] = df.DateTime1['Date_Time'].dt.day
df.DateTime1['Hour'] = df.DateTime1['Date_Time'].dt.hour
```

```python
df.DateTime1.tail()
```

```python
df.DateTime['Date_And_Time'] = df.DateTime1['Date_Time']
df.DateTime['Year'] = df.DateTime1['Year']
df.DateTime['Month'] = df.DateTime1['Month']
df.DateTime['Day'] = df.DateTime1['Day']
df.DateTime['Hour'] = df.DateTime1['Hour']
df.DateTime.head()
```

## New Table

```python
New_df = df.DateTime[['Date_And_Time', 'Year', 'Month', 'Day', 'Hour', 'Sensor_Name', 'Hourly_Counts']]
New_df.head(25)
```

```python
plt.clf()
New_df['Date_And_Time'].map(lambda d: d.month).plot(kind='hist')
plt.show()
```

```python
plt.clf()
New_df['Date_And_Time'].map(lambda d: d.hour).plot(kind='hist')
plt.show()
```
